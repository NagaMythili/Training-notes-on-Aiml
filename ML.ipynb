```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r'E:\Salary_Data.csv')

#creates independable variable
X=df.iloc[:, :-1].values

Y=df.iloc[:,1].values
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()

# passing the data using the fit method
reg.fit(X_train,Y_train)
#predicting the test set results
y_pred=reg.predict(X_test)
print(y_pred)
plt.scatter(X_train,Y_train,color='red')
plt.plot(X_test,reg.predict(X_test),color='indigo')
#plt.plot(X_train,reg.predict(X_train),color='indigo')
plt.show()
#print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
    'Result':pd.Series([100,90,50,70,80])}
df=pd.DataFrame(ds)
#print(df)
X= df[['SUB1', 'SUB2']].values
# Labels (Y) - Predicting 'ROLL NO' based on 'SUB1' and 'SUB2'
Y = df['Result'].values
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
# Create an instance of the model
model = LinearRegression()
# Fit the model
model.fit(X_train, Y_train)
y_pred=model.predict(X_test)
print(y_pred)
plt.scatter(Y_test, y_pred,color='red')
plt.xlabel('Actual Result')
plt.ylabel('Predicted Result')
plt.title('Actual vs Predicted Results')
plt.show()
```
