# BLOB :    stands for Binary Large Object
    is basically a collection/ chunk of binary data stored in the form of a single entity in a DB (database) system. 

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

#Load Spacy model and add the SpacyTectblob component
nlp=spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
def analyze_sentiment(text):
    
    try:
        doc=nlp(text)
        #access the sentiment polarity from spaCyTextBlob
        sentiment_polarity = doc._.polarity
        #classify sentiment based on polarity
        if sentiment_polarity >0:
            return "Positive"
        elif sentiment_polarity==0:
            return "Neutral"
        else:
            return "Negative"
    except Exception as e:
        print(f"Error:{e}")
        return "Error"
#example usage:
text_to_analyze="Hi !!Welcome"
res=analyze_sentiment(text_to_analyze)
print(f"Sentiment Analysis Result: {res}")
```
